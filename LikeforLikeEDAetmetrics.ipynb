```python
#Source elaborée par Yacine, reprend toutrs les données des magazins Like for Like ouverts sans discontinuité
from google.colab import auth
import pandas as pd

# Authenticate your Google Colab environment
auth.authenticate_user()

# Define your SQL query
query = '''
    SELECT
        date_date,
        shop_id,
        product_cat,
        turnover,
        pieces,
        receipts,
        Like_for_like
    FROM `solar-semiotics-393108.to_the_moon.open_shop_all_data`
    ORDER BY date_date
'''

# Read the data from BigQuery into a DataFrame
df = pd.read_gbq(query, project_id="solar-semiotics-393108")

# Now, 'df' contains the results of your SQL query
```

```python
# Print rows with null or negative values in the 'turnover' column
filtered_df = df2[(df2['turnover'].isnull()) | (df2['turnover'] < 0)]
print(filtered_df)
```

```python
negative_rows_pieces = df2[df2['pieces'] < 0]
negative_rows_pieces
```

```python
#Delete null and negative values rows (turnover & pieces)and generate new df5 dataframe (df3 and df4 apply the replace by mean value method)
df5 = df2[df2.min(axis=1) >= 0]
df5
```

```python
# Calculate the mean of non-null and positive values in the "turnover" column and generate df3 new dataframe
mean_turnover = df2[df2['turnover'] >=0]['turnover'].mean()

# Replace null or negative values with the mean
df2['turnover'].fillna(mean_turnover, inplace=True)

# Create a new DataFrame with the replacements
df3 = df2.copy()

# Now, 'new_df' contains the DataFrame with null or negative values in the "turnover" column replaced by the mean
print(df3)
```

```python
# Calculate the mean of non-null and positive values in the "pieces" column  and keep it as df3 dataframe
mean_pieces = df3[df3['pieces'] >= 0]['pieces'].mean()

# Replace null or negative values with the mean
df3['pieces'].fillna(mean_pieces, inplace=True)
df3['pieces'] = df3['pieces'].apply(lambda x: mean_pieces if x < 0 else x)

# Create a new DataFrame with the replacements
df4 = df3.copy()

# Now, 'new_df' contains the DataFrame with null or negative values in the "pieces" column replaced by the mean
print(df4)
```

```python
#df5 = df2.drop(df2[df2['col1'] < 0].index, inplace=True)
df5 = df2[df2.min(axis=1) >= 0]
df5
```

```python
#Apply groupby by shop_id and date to mix all products categories
aggregeturnover_df = df5.groupby(['shop_id','date_date']) [['turnover',"pieces","receipts"]].sum().reset_index()
aggregeturnover_df
```

```python
#calculate average basket value from cleaned data
aggregeturnover_df['average_basket'] = aggregeturnover_df['turnover']/aggregeturnover_df['receipts']
aggrege_df= aggregeturnover_df
aggrege_df
```

```python
#Find number of items per receipt or sales index
sales_index = aggrege_df['pieces']/aggrege_df['receipts']
aggrege_df['sales_index'] = sales_index
aggrege_df
```
